```python
from skimage import data
```

```python
image = data.astronaut()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image)
plt.show()
```

```python
from skimage import segmentation
```

```python
unsupervised_segment_image = segmentation.slic(image,n_segments=400)
```

```python
from skimage.color import label2rgb
labeled_image = label2rgb(unsupervised_segment_image)
```

```python
plt.imshow(labeled_image)
plt.show()
```

```python
# get the image contours
# steps
# 1 - convert the given image into gray image
# 2 - generate a threshold value 
# 3 - apply binary thrshold value
# 4- use the find_contour method to get the contours border line.
from skimage.color import rgb2gray
gray_image = rgb2gray(image)
plt.imshow(gray_image)
plt.show()
```

```python
from skimage.filters import threshold_otsu
value = threshold_otsu(gray_image)
print(value)
```

```python
# applying the value
threshold_value = gray_image>value
print(threshold_value)
```

```python
# contours
from skimage.measure import find_contours
measure = find_contours(threshold_value)
```

```python
print(measure)
```

```python
plt.imshow(measure)
plt.show()
```
