```python
import skimage
import matplotlib.pyplot as plt
import numpy as np
```

```python
cam_img = skimage.data.camera()
```

```python
# using Edge function (sobel)
from skimage.filters import sobel
# sobel function is used to detect the edges of the object
cam_edges = sobel(cam_img)
plt.imshow(cam_edges)
plt.axis("off")
```

```python
# for smoothing we use Gaussian function
from skimage.filters import gaussian
gass_img = gaussian(cam_img,)
plt.imshow(gass_img)
plt.axis("off")
```

```python
cat_img = skimage.data.cat()
plt.imshow(cat_img)
plt.axis("off")
```

```python
from skimage.filters import sobel
from skimage.color import rgb2gray
gray_img = rgb2gray(cat_img)
cat_edge = sobel(gray_img)
plt.imshow(cat_edge)
plt.axis("off")
```

```python
plt.hist(cat_edge)
```
