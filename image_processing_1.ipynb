```python
# basically there are two types of thresholding 
# 1- Global Thresholding -> used for b
# 2- Local Thresholding

camera_img = ski.data.camera()
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu,threshold_local
Local_binary_thresh = threshold_local(camera_img)
otsu_binary_thresh = threshold_otsu(camera_img)
print(f"local thresh value is : {Local_binary_thresh} and Otsu thresh value is {otsu_binary_thresh}")
```

```python
# Working on above thresh Value to apply on an image
Appl_thresh_value = camera_img<=Local_binary_thresh
Appl_thresh_value_2 = camera_img>=Local_binary_thresh
images = [ Appl_thresh_value , Appl_thresh_value_2]
for i in range(2):
    plt.subplot(1,2,i+1)
    show_image(images[i])
```

```python
# working on Otsu Thresholding
app_otsu_thresh = camera_img<=otsu_binary_thresh
app_otsu_thresh_2 =  camera_img>=otsu_binary_thresh
otsu_Threshold_images = [app_otsu_thresh,app_otsu_thresh_2]
Threshold_type = ["ForeGround","BackGround"]
for j in range(2):
    plt.subplot(2,1,j+1)
    plt.title(Threshold_type[j])
    show_image(otsu_Threshold_images[j])
```

```python
## using Transformation
import skimage
from skimage.transform import rotate
from skimage.data import camera
img = camera()
rotate_img = rotate(img,90)
plt.imshow(rotate_img)
plt.axis("off")
```

```python
# using morphology
# that is used on dimension of an image like add pixel to the image'object such kind of processes is known as dilation and erosion is pixel remove from boundry of object in an image

from skimage import morphology
# these operation are suitable for binary images(threshold images)
img =  skimage.data.horse()
# dilation
img_dilation = morphology.binary_dilation(img)
mix_images = [img,img_dilation]
labels =  ["original","Binary_dilated"]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(mix_images[i])
    plt.title(labels[i])
    plt.axis("off")
```

```python
erosion_img = morphology.binary_erosion(img)
mix_images = [img,erosion_img]
labels =  ["original","Binary_Erosion"]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(mix_images[i])
    plt.title(labels[i])
    plt.axis("off")
```
